#### Расчет Retention Rate

```python
first_activity_date = visits.groupby(['uid'])['start_ts'].min()
first_activity_date.name = 'first_activity_date'
visits = visits.join(first_activity_date,on='uid')
visits['first_visit_month'] = visits['first_activity_date'].astype('datetime64[M]')
visits['visit_month'] = visits['date'].astype('datetime64[M]')
```

Нашли дату первого посещения у каждого пользователя

```python
visits['cohort_lifetime'] = (
    visits['visit_month'] - visits['first_visit_month']
)

visits['cohort_lifetime'] = visits[
    'cohort_lifetime'
] / np.timedelta64(1, 'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round()
visits['cohort_lifetime'] = visits['cohort_lifetime'].astype(
    'int'
) 
```

Сформировали когорты по месяцу первого посещения

```python
cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
```

```python
cohorts
```

```python
initial_user_count = cohorts[cohorts['cohort_lifetime']==0][['first_visit_month','uid']]
initial_user_count = initial_user_count.rename(columns={'uid': 'cohort_uid'})
initial_user_count
```

Нашли начальное количество пользователей в когорте

```python
cohorts = cohorts.merge(initial_user_count, on = 'first_visit_month')
```

```python
cohorts['retention'] = (cohorts['uid'] / cohorts['cohort_uid'])
```

```python
cohorts['retention']
```

Расчитали коэффициент удержания

```python
retention_pivot = cohorts.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
) 
retention_pivot
```

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='black'); 
```

Построил тепловую карту для Retention Rate по когортам

```python
retention_pivot[2].mean()
```

Средний коэффициент удержания 5%

#### Расчет LTV

```python
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']
first_orders
```

Находим месяц первой покупки

```python
cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

```python
orders_new = pd.merge(orders,first_orders, on='uid')
orders_new
```

Создаем таблицу когорт и объеденяем таблицу orders с таблицей даты первой покупки

```python
orders_new
```

```python
cohorts = (
    orders_new.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
```

```python
cohorts
```

Формируем когорты по месяцам, складывая их суммы заказа

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
```

```python
report
```

Объеденяем таблицу с количество людей в когорте с датами и суммами

```python
margin_rate = 1

report['gp'] = report['revenue'] * margin_rate
report['age'] = (
    report['order_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
```

```python
report
```

Находим возраст когорг

```python
report['ltv'] = report['gp'] / report['n_buyers']

output = report.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
)

output.fillna('')
```

Расчитали LTV для всех когорт

```python
ltv = output.cumsum(axis=1)
ltv.fillna('')
```

```python
ltv = ltv[ltv.loc[:,5].notna()]
ltv.mean()
```

Расчитал средний LTV по когортам за полгода Можно заметить, что ltv в когортах увеличивается со временем

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('LTV')
sns.heatmap(ltv, annot=True, fmt='.3', linewidths=1, linecolor='black'); 
```
